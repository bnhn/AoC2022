```python
from operator import add,sub,truediv,mul

res,res2=0,0
M=dict()
calc={'+':add,'-':sub,'*':mul,'/':truediv}

with open('Data/D11D.txt') as file:
    m,n=0,0
    
    for line in file:
        if line.strip()=='':
            continue
            
        if line[0]=='M':
            M[m]={
                'old':[],
                'new':[],
                'opr':[],
                'test':0,
                't1':0,
                't2':0,
                'ins':0
            }
            m+=1
            continue
        
        line=line.strip()
        l=line.split(':')
        l0,l1=l[0].strip(),l[1].strip()
        
        if n==4:
            M[m-1]['t2']=int(l1[len(l1)-1].strip())
            n=0
            continue
        elif n==3:
            M[m-1]['t1']=int(l1[len(l1)-1].strip())
        elif n==2:
            M[m-1]['test']=int(l1.split()[2].strip())
        elif n==1:
            M[m-1]['opr']=l1.split('=')[1].strip().split()
        elif n==0:
            o=[int(_.strip()) for _ in l1.split(',')]
            M[m-1]['old']=o
        n+=1
        

for i in range(20):
    for mon in M:
        old=M[mon]['old']
        M[mon]['new']=[]
        opr=M[mon]['opr'][1]
        a=M[mon]['opr'][0]
        b=M[mon]['opr'][2]
        test=M[mon]['test']
        t1=M[mon]['t1']
        t2=M[mon]['t2']
        
        if not old:
            continue

        for _ in range(len(old)):
            M[mon]['ins']+=1
            oo=old.pop(0)
            if b == 'old':
                c=calc[opr](oo,oo)
            else:
                c=calc[opr](oo,int(b))
            c=c//3
            M[mon]['new'].append(c)

            #check monkey throw destination
            if c%test==0:
                M[t1]['old'].append(c)
            else:
                M[t2]['old'].append(c)
        
mb=[]
for i in M:
    mb.append(M[i]['ins'])
    print(M[i]['old'],M[i]['ins'])
mb.sort(reverse=True)
res=mb[0]*mb[1]
print(mb)
print(res)
```
