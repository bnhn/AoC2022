```python
import time

start=time.time()

def parse_data(data_path):
    M={}
    with open(data_path) as file:
        m,n=0,0
    
        for line in file:
            if line.strip()=='':
                continue

            if line[0]=='M':
                M[m]={
                    'old':[],
                    'new':[],
                    'opr':[],
                    'test':0,
                    't1':0,
                    't2':0,
                    'ins':0
                }
                m+=1
                continue

            line=line.strip()
            l=line.split(':')
            l0,l1=l[0].strip(),l[1].strip()

            if n==4:
                M[m-1]['t2']=int(l1[len(l1)-1].strip())
                n=0
                continue
            elif n==3:
                M[m-1]['t1']=int(l1[len(l1)-1].strip())
            elif n==2:
                M[m-1]['test']=int(l1.split()[2].strip())
            elif n==1:
                M[m-1]['opr']=l1.split('=')[1].strip().split()
            elif n==0:
                o=[int(_.strip()) for _ in l1.split(',')]
                M[m-1]['old']=o
            n+=1
    return M

def monkey_business(MM,rounds,part_one=True,super_mod=0):
    mb=[]
    for i in range(rounds):
        for mon in MM:
            old=MM[mon]['old']
            MM[mon]['new']=[]
            opr=MM[mon]['opr'][1]
            a=MM[mon]['opr'][0]
            b=MM[mon]['opr'][2]
            test=MM[mon]['test']
            t1=MM[mon]['t1']
            t2=MM[mon]['t2']

            if not old:
                continue

            for _ in range(len(old)):
                MM[mon]['ins']+=1
                oo=old.pop(0)
                if b == 'old':
                    c=calc[opr](oo,oo)
                else:
                    c=calc[opr](oo,int(b))
                if part_one:
                    c=c//3
                else:
                    c%=super_mod
                MM[mon]['new'].append(c)

                #check monkey throw destination
                if c%test==0:
                    MM[t1]['old'].append(c)
                else:
                    MM[t2]['old'].append(c)
    
    for i in MM:
        mb.append(MM[i]['ins'])
        mb.sort()
    return mb
```
